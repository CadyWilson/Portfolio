## Investigation Overview


> I took the approach of a consumer while analyzing this dataset. How might different social constructs like income, credit scores etc. affect the interest rate of a loan. Interest is what you will be paying on top of the loan making it more expensive. The higher the interest, the higher the payback amount.


## Dataset Overview

> This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. After Dropping duplicate rows and condensing down to only key variables I ended with just 21 features and 113,066 rows.

Key Features included:
* `BorrowerAPR`: The Borrower's Annual Percentage Rate (APR) for the loan.
* `ProsperRating`: The Prosper Rating assigned at the time the listing was created between AA - HR. The Order being AA, A, B, C, D, E and HR
* `DebtToIncomeRatio`: The debt to income ratio of the borrower at the time the credit profile was pulled. The ration being how much debt you have compared to how much income you receive. (capped at 10%)
* `IncomeRange`: A range of 6 buckets of the income of the borrower.

## Distribution of Income Ranges and Credit Ratings

> High credit scores correlate with higher incomes, as anticipated. The plot below shows that borrowers of any income level can recieve a loan. The income groups that recieved the majority of the loans are typically classed as lower and lower-middle class. (According to Pew Research Center)

## Loan Amount per Employment Status

> Prosper doesn't tend to give out loans larger than around twenty thousand dollars to borrowers of any employment status. It is interesting to note that employment status doesn't factor nearly as much into loan amount or the number of loans under five thousand dollars given as the other variables do.

## Loan Term vs Credit Rating

> Unsuprisingly, across all loan term length there is a negaitve correlation between the amount given and credit rating. Borrowers with worse credit ratings are higher risk, and to give them a larger amount for a loan would not be worth the risk.

> There is a positive correlation between higher credit ratings and both Borrower APR and their Debt To Income Ratio in all three loan term lengths. That is, no matter the length of the loan, a better credit score will yield you a better interest rate, and the higher the likelihood they have a low debt to income ratio. Making the borrower from the perspective of Prosper, lower risk.



## Credit vs APR

>There is a clear negative correlation between credit rating and interest rate. The lower your credit score, the higher your interest rate. Meaning in the end you will pay more overall for your loan.

> It does appear that prosper may have a system where APR is fixed to credit rate, as the steps down by rate are almost uniform.
