# An Analysis of We Rate Dogs, the Twitter account sweeping the nation.

## Becuase they're good dogs, Brent.

As a data wrangling project for a Udacity account I’m working on, I was tasked with cleaning and wrangling a LOT of data related to @dog_rates. The Twitter account that gives ridiculous ratings to dogs for no other reason than it can. After a ton of time and effort in the cleaning and wrangling phase (making that data make sense and look nice) I was finally able to do the fun part. Analyzing all that sweet, clean data.

I started off with the boring stuff, post engagement over time. As we can see here, the account only grew in popularity over time. I did only have the data up until July 0f 2017, but that was enough to see a very positive upward trend in both likes and retweets.


Next up, the fun stuff. We Rate Dogs posts pictures of the dogs they are rating. Using my data and the results of a neural network built by Udacity as well, I got to the good stuff. Best breed and top dog name listed in the tweets. The winner of the breeds are exactly what you’d expect. Golden retrievers, closely followed by labrador retrievers.

And, to wrap it all up. The best of the best. The most favorited and the most retweeted dogs on the account.
